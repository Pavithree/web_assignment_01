```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
SIN = []
COS = []
random_values = []
random.seed(42)

#generating 100 random numbers in the range(0-1000)
for number in range(0,100):
    rand_num = random.uniform(0,1000)
    random_values.append(rand_num)
    #Converting the random number to radian
    degree_to_radian = np.deg2rad(rand_num)
    #Finding SIN value of the converted random number
    SIN.append(np.sin(degree_to_radian))
    #Finding COS value of the converted random number
    COS.append(np.cos(degree_to_radian))

print(SIN)
print(COS)
```

```python
plt.scatter(random_values,SIN,color='blue',label='SIN values')
plt.scatter(random_values,COS,marker='+', color='red',label='COS values')
plt.xlabel("Random number")
plt.ylabel("Sin and cos values")
plt.title("Sine and cosine values")
plt.legend(loc='upper right')
plt.show()
```
