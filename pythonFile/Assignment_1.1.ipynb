```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
SIN = []
COS = []
random_values = []
random.seed(42)

#generating 100 random numbers in the range(0-1000)
for number in range(0,100):
    rand_num = random.uniform(0,1000)
    random_values.append(rand_num)

#sorting the values in order to get smooth sin and cos curves
random_values_sorted = sorted(random_values)

#converting degrees to radians
for number in random_values_sorted:
    SIN.append(np.sin(np.radians(number)))
    COS.append(np.cos(np.radians(number)))
print(SIN)
print(COS)
```

```python
plt.scatter(random_values_sorted,SIN,color='blue',label='SIN values')
plt.scatter(random_values_sorted,COS,marker='+', color='red',label='COS values')
plt.xlabel("Random number")
plt.ylabel("Sin and cos values")
plt.title("Sine and cosine values")
plt.legend(loc='upper right')
```
