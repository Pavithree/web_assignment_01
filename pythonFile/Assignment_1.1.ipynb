```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
SIN = []
COS = []
random_values = []
random.seed(42)
for number in range(0,100):
    rand_num = random.uniform(0,1000)
    random_values.append(rand_num)
    SIN.append(np.sin(rand_num))
    COS.append(np.cos(rand_num))
print(SIN)
print(COS)
print(random_values)
```

```python
plt.scatter(random_values,SIN,color='blue',label='SIN values')
plt.scatter(random_values,COS,marker='+', color='red',label='COS values')
plt.xlabel("Random number")
plt.ylabel("Sin and cos values")
plt.title("Sine and cosine values")
plt.legend(loc='upper right')
```
